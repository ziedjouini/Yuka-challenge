# Mission 4

Quel est le nombre de colonne appartenant à chaque quartile des taux de remplissage ? Autrement dit, combien de colonne sont remplies à 0%, entre 0 et 25%, entre 25 et 50%...

```
mylist0=[]
mylist25=[]
mylist50=[]
mylist75=[]
mylist100=[]
for taux in dataframe(df_object):
  if taux == 0:
    mylist0.append(taux)
  elif 0<taux <=25:
    mylist25.append(taux)
  elif  25<taux <=50:
    mylist50.append(taux)
  elif 50<taux<=75:
    mylist75.append(taux)
  else:
    mylist100.append(taux)

print("nombre de colonnes dont le taux est egale à 0 est de ", len(mylist0))
print("nombre de colonnes dont le taux est inférieur ou egale à 25 est de ", len(mylist25))
print("nombre de colonnes dont le taux est inférieur ou egale à 50 est de ", len(mylist50))
print("nombre de colonnes dont le taux est inférieur ou egale à 75 est de ", len(mylist75))
print("nombre de colonnes dont le taux est inférieur ou egale à 100 est de ", len(mylist100))
  
```
