```
def missing_values():
  return df.isna().sum()/len(df)
```

```
def column_fill_rate(df):
    for col in df.columns:
        fill_rate = (df[col].count() / len(df)) * 100
        print(f'{col} : {fill_rate:.2f}%')
```

```
column_fill_rate(df)
```

```
def column_fill(df):
    for col in df.columns:
        fill_rate = (df.notnull().sum() / len(df)) * 100
        #print(f'{col} : {fill_rate:.2f}%')
        return fill_rate
```
